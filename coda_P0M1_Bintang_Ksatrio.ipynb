```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#Referensi HTML
#Class Product
#<div class="bl-flex-item mb-8">

#Nama Product
#<p class="bl-text bl-text--body-14 bl-text--secondary bl-text--ellipsis__2">

#Harga Product
# <p class="bl-text bl-text--semi-bold bl-text--ellipsis__1 bl-product-card-new__price">

#Rating Product
# <p class="bl-text bl-text--caption-12 bl-text--bold">

# Store Name
# <p class="bl-text bl-text--caption-12 bl-text--secondary bl-text--ellipsis__1 bl-product-card-new__store-name"

#Store Loc
#<p class="bl-text bl-text--caption-12 bl-text--secondary bl-text--ellipsis__1 bl-product-card-new__store-location">
```

```python
# Wadah untuk menampung Value
names = []
prices = []
ratings = []
stores = []
locs = []

# variabel untuk menyimpan webdriver
driver = webdriver.Edge()

#Looping untuk mengambil data dari halaman 1 - 5
for page in range (1,6): 
    url = f"https://www.bukalapak.com/products?page={page}&search%5Bkeywords%5D=basreng"
    
    driver.get(url)
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # Looping untuk mencari data yg ada pada class box produk
    for data in soup.find_all('div',{'class':'bl-flex-item mb-8'}):
        try:
            nama = data.find('p', {'class':'bl-text bl-text--body-14 bl-text--secondary bl-text--ellipsis__2'}).get_text()# mengambil semua data nama produk
            names.append(nama.strip().replace('\n','')) # Menghilangkan value yg tidak diinginkan
        except:
            names.append(None) #diisi none jika tidak ada data
        
        try:
            harga = data.find('p',{'class':'bl-text bl-text--semi-bold bl-text--ellipsis__1 bl-product-card-new__price'}).get_text()# mengambil semua data harga produk
            prices.append(harga.replace('\n','').replace(' ','').replace('.','')) # Menghilangkan value yg tidak diinginkan
        except:
            prices.append(None) # diisi none jika tidak ada data

        try:
            rate = data.find('p',{'class':'bl-text bl-text--caption-12 bl-text--bold'}).get_text() # mengambil semua data rating produk
            ratings.append(rate.replace('\n','').replace(' ',''))# Menghilangkan value yg tidak diinginkan
        except:
            ratings.append(None) # diisi none jika tidak ada data

        try:
            toko = data.find('p',{'class':'bl-text bl-text--caption-12 bl-text--secondary bl-text--ellipsis__1 bl-product-card-new__store-name'}).get_text() # mengambil semua data Nama Toko
            stores.append(toko.replace('\n','').replace(' ',''))# Menghilangkan value yg tidak diinginkan
        except:
            stores.append(None) # diisi none jika tidak ada data

        try:
            lokasi = data.find('p',{'class':'bl-text bl-text--caption-12 bl-text--secondary bl-text--ellipsis__1 bl-product-card-new__store-location'}).get_text() # mengambil semua data Nama Toko
            locs.append(lokasi.replace('\n','').replace(' ',''))# Menghilangkan value yg tidak diinginkan
        except:
            locs.append(None) # diisi none jika tidak ada data

# tutup driver setelah berhasil melakukan looping
driver.quit()
        
# melihat panjang data yg berhasil didapat
len(names),len(prices),len(ratings),len(stores),len(locs)
```

```python
# Simpan value yg berhasil di scrape ke dalam dataframe df
data = ({
    'Product_Names': names,
    'Prices' : prices,
    'Ratings' : ratings,
    'Store_Names' : stores,
    'Location' : locs
})

df = pd.DataFrame(data)
# tampilkan dataframe
df.head()
```

```python
# Tampilkan informasi mengenaik dataframe df
df.info()
```

```python
# Ubah tipe data prices menjadi int dan rating menjadi float
df['Prices'] = df['Prices'].astype(int)
df['Ratings'] = df['Ratings'].astype(float)
df.info()
```
